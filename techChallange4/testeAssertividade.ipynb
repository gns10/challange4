# Importando dados

# Featuring engineering

## Arredondamento dos dados numéricos

```python
cols = ['Age','Weight','FCVC','NCP','CH2O','FAF','TUE']
df[cols] = df[cols].round().astype(int)
df
```

## Criando coluna IMC

```python
df['IMC'] = df['Weight']/(df['Height'] ** 2)
df['IMC'] = df['IMC'].round().astype(int)
df
```

```python
df['Obesity'].unique()
```

# Avaliando colunas string

* family_history yes: 1 no: 0
* FAVC yes: 1 no: 0
* FCVC yes: 1 no: 0
* CAEC no: 0 Sometimes: 1 Frequently: 2 Always: 3
* Smoke yes:1 no:0
* SCC yes: 1 no: 0
* CALC no: 0 Sometimes: 1 Frequently: 2 Always: 3
* MTRANS Automobile: 0 Motorbike: 1 Bike 2 Public_Transportation 3 Walking 4


```python
df.columns
```

```python
df_num = df.copy()

#Mapeamentos
map_gender = {'Female': 0, 'Male':1}
map_family_history = {'yes':1, 'no':0}
map_FAVC = {'yes':1, 'no':0}
map_CAEC = {'no':0,'Sometimes':1, 'Frequently':2 ,'Always':3}
map_smoke = {'yes':1, 'no':0}
map_SCC = {'yes':1, 'no':0}
map_CALC = {'no':0,'Sometimes':1, 'Frequently':2 ,'Always':3}
map_METRANS = {'Automobile': 0, "Motorbike": 1, 'Bike':2, 'Public_Transportation': 3, 'Walking': 4}

df_num['Gender'] = df_num['Gender'].map(map_gender)
df_num['family_history'] = df_num['family_history'].map(map_family_history)
df_num['FAVC'] = df_num['FAVC'].map(map_FAVC)
df_num['CAEC'] = df_num['CAEC'].map(map_CAEC)
df_num['SMOKE'] = df_num['SMOKE'].map(map_smoke)
df_num['SCC'] = df_num['SCC'].map(map_SCC)
df_num['CALC'] = df_num['CALC'].map(map_CALC)
df_num['MTRANS'] = df_num['MTRANS'].map(map_METRANS)

df_num.info()

```

# Criando o modelo

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Target
y = df_num['Obesity']

# Ajustando a string (ex.: 'Obese', 'Normal', etc.) para classes [0, 1, 2, 3, 4, 5, 6]
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Features (incluindo IMC e demais colunas)
X = df_num.drop(columns=['Obesity'])
```

```python
#Divisão treino e teste
# Teste com 30%
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)
```

```python
# Escolhendo o random forest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Acurácia:", acc)
```

```python
# Adotar o F1 score para avaliar o modelo
from sklearn.metrics import accuracy_score, f1_score

# Treinar modelo
model.fit(X_train, y_train)

# Previsões
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Métricas
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

train_f1 = f1_score(y_train, y_train_pred, average='macro')
test_f1 = f1_score(y_test, y_test_pred, average='macro')

print("=== Accuracy ===")
print("Treino:", train_acc)
print("Teste :", test_acc)

print("\n=== F1 Macro ===")
print("Treino:", train_f1)
print("Teste :", test_f1)
```

O modelo apresenta leve overfitting esperado para RandomForest, porém a diferença entre treino e teste é inferior a 5 p.p., indicando boa capacidade de generalização.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    model,
    X,
    y_encoded,
    cv=skf,
    scoring='f1_macro'
)

print("F1 por fold:", cv_scores)
print("Média CV:", np.mean(cv_scores))
print("Desvio padrão:", np.std(cv_scores))
```

O modelo apresentou F1 médio de 0.975 (±0.008) em validação cruzada estratificada 5-fold, indicando elevada estabilidade e baixa variância. O gap de generalização em relação ao conjunto de treino foi inferior a 3 p.p., não caracterizando overfitting relevante.

```python
# Função interativa para input de dados

def testar_paciente(model, le):
    print("=== Teste de Predição de Obesidade ===")

    # Inputs do usuário
    gender = int(input("Gênero (0=feminino, 1=masculino): "))
    age = int(input("Idade: "))
    height = float(input("Altura (m): "))
    weight = int(input("Peso (kg): "))

    # Calcula IMC automaticamente
    imc = float(weight / (height ** 2))

    print(f"IMC calculado automaticamente: {imc}")

    dados = {
        'Gender': gender,
        'Age': age,
        'Height': height,
        'Weight': weight,
        'family_history': int(input("Histórico familiar (1=sim, 0=não): ")),
        'FAVC': int(input("Consome comida calórica frequentemente? (1=sim, 0=não): ")),
        'FCVC': int(input("Come vegetais frequentemente? (1=sim, 0=não): ")),
        'NCP': int(input("Número de refeições principais por dia (1 a 4): ")),
        'CAEC': int(input("Come entre refeições? (0=Não, 1=Às vezes, 2=Freq., 3=Sempre): ")),
        'SMOKE': int(input("Fuma? (1=sim, 0=não): ")),
        'CH2O': int(input("Consumo de água por dia (1=baixo, 2=médio, 3=alto): ")),
        'SCC': int(input("Monitora calorias? (1=sim, 0=não): ")),
        'FAF': int(input("Atividade física semanal (0 a 3): ")),
        'TUE': int(input("Tempo usando tecnologia por dia (0 a 2): ")),
        'CALC': int(input("Consumo de álcool? (0=Não, 1=Às vezes, 2=Freq., 3=Sempre): ")),
        'MTRANS': int(input("Transporte (0=Carro,1=Moto,2=Bicicleta,3=Transp.Público,4=Caminhada): ")),
        'IMC': imc
    }

    df_paciente = pd.DataFrame([dados])

    pred = model.predict(df_paciente)[0]
    classe = le.inverse_transform([pred])[0]

    print("\n=== Resultado ===")
    print(f"Classe de peso prevista: {classe}")
```

## Histórico familiar é um forte preditor ?

```python


# Calcula proporções de obesidade por histórico familiar
prop = (
    df.groupby(["family_history", "Obesity_label"])
      .size()
      .groupby(level=0)
      .apply(lambda x: x / x.sum())
      .reset_index(name="Proportion")
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=prop,
    x="family_history",
    y="Proportion",
    hue="Obesity_label",
    palette="magma",
    hue_order=order
)

plt.title("Proporção das Classes de Obesidade por Histórico Familiar", fontsize=14)
plt.xlabel("Histórico Familiar de Excesso de Peso")
plt.ylabel("Proporção dentro do Grupo")
plt.legend(title="Obesidade")
plt.tight_layout()

# removendo eixos superior e o direito
sns.despine(top=True, right=True)

plt.show()
```

## Hábitos alimentares são determinates para a obesidade ?

```python
# Ajustando as legendas do  FAVC
favc_labels = {"yes": "Sim", "no": "Não"}

#Invertendo a ordem da lista
order_invert = order[::-1]

prop = (
    df.groupby(["Obesity_label", "FAVC"])
      .size()
      .groupby(level=0)
      .apply(lambda x: x / x.sum())
      .reset_index(name="Proportion")
)

plt.figure(figsize=(12, 6))

sns.barplot(
    data=prop,
    x="Obesity_label",
    y="Proportion",
    hue="FAVC",
    palette="deep",
    order=order_invert
)

plt.title("Frequência no consumo de alimentos Calóricos", fontsize=14)
plt.xlabel("Classe de Obesidade")
plt.ylabel("Proporção dentro da Classe")
plt.xticks(rotation=0, ha="center")

# Ajustar legenda
handles, labels = plt.gca().get_legend_handles_labels()
new_labels = [favc_labels[l] for l in labels]
plt.legend(handles, new_labels, title="Consumo de Alimentos Calóricos")

sns.despine(top=True, right=True)
plt.tight_layout()
plt.show()
```

## Comportamentos impactam na obesidade

```python
# CALC - (Alcohol consumption
# CH2O — (Daily water consumption)
# FAF — (Physical activity frequency)
# TUE — (Time using electronic devices)
# NCP — (Number of main meals)

# Criar tabela para heatmap
heatmap_data = df.groupby("Obesity_label")[["CALC", "CH2O", "FAF", 'TUE', 'NCP']].mean()
heatmap_data = heatmap_data.reindex(order_invert)

plt.figure(figsize=(12, 6))

sns.heatmap(
    heatmap_data.T,
    annot=True,
    cmap="YlOrRd",
    linewidths=0.5,
    cbar_kws={"label": "Média do Comportamento"}
)

plt.title("Heatmap dos Comportamentos (CALC, SMOKE, FAF) por Classe de Obesidade", fontsize=14)
plt.xlabel("Classe de Obesidade")
plt.ylabel("Comportamento")

plt.tight_layout()
plt.show()
```
